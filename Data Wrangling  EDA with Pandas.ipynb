# CODING TASK #7. PANDAS WITH FUNCTIONS

```python
# Let's read a CSV file using Pandas as follows
hr_df = pd.read_csv('Human_Resources.csv')
hr_df.head()
```

```python
hr_df = hr_df.dropna()
hr_df.head()
```

```python
rate = 1.10
def net_degeri_artir(hr_df):
    yeni_net_deger = hr_df * rate  # %10 artırma işlemi
    return yeni_net_deger

# DataFrame'e işlevi uygulayalım
hr_df['MonthlyIncome'] = hr_df['MonthlyIncome'].apply(net_degeri_artir)
hr_df['DailyRate'] = hr_df['DailyRate'].apply(net_degeri_artir)
hr_df['HourlyRate'] = hr_df['HourlyRate'].apply(net_degeri_artir)
hr_df['MonthlyRate'] = hr_df['MonthlyRate'].apply(net_degeri_artir)
```

```python
hr_df.head()
```

**PRACTICE OPPORTUNITY #7:**
- **Define a function that doubles the DailyRate and adds $100**
- **Apply the function to the DataFrame**
- **Calculate the updated total Daily Rate of all employees combined**

```python
#Define a function that doubles the DailyRate and adds $100
#Apply the function to the DataFrame
def daily_rate_artır(hr_df):
    yeni_net_deger = (hr_df * 2) + 100  # iki katının 100 fazlası
    return yeni_net_deger

# DataFrame'e işlevi uygulayın
hr_df['DailyRate'] = hr_df['DailyRate'].apply(daily_rate_artır)
hr_df.head(5)
```

# CODING TASK #8: HISTOGRAM AND CORRELATION

- **Several features such as 'MonthlyIncome' and 'TotalWorkingYears' are tail heavy**
- **Job level is strongly correlated with total working hours**
- **Monthly income is strongly correlated with Job level**
- **Monthly income is strongly correlated with total working hours**
- **Age is strongly correlated with monthly income**

```python
columns_to_drop = ["EmployeeCount", "StandardHours"]
hr_df1=hr_df.drop(columns=columns_to_drop, axis=1)
plt.figure(figsize=(15, 12))
sns.heatmap(hr_df1.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, annot_kws={'size': 12})  # Yazıları büyütelim
plt.title('Heat Map')
plt.show()
```

```python
#Several features such as 'MonthlyIncome' and 'TotalWorkingYears' are tail heavy
#Job level is strongly correlated with total working hours
#Monthly income is strongly correlated with Job level
#Monthly income is strongly correlated with total working hours
#Age is strongly correlated with monthly income

strong_corr = hr_df[['MonthlyIncome', 'TotalWorkingYears','JobLevel', 'Age']]

corr_matrix = strong_corr.corr()
# Korelasyon matrisini kullanarak bir heatmap çizelim
plt.figure(figsize=(4, 3))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Strong Correlation Relationship')
plt.show()
```
